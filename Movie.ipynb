```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("movies.csv")
df.head()
```

```python
df.shape
```

```python
# Let's see if there i any missing data

for col in df.columns:
    missing = np.mean(df[col].isna())
    print('{} - {}%'.format(col, missing))
```

```python
# Data Types for our columns

df.dtypes
```

```python
# Filling the missing value with their median
df.isna().sum()
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
# Filling the missing values

df["rating"] = df["rating"].fillna(method = 'pad')
df["released"] = df["released"].fillna(method = 'pad')
df["score"] = df["score"].fillna(method = 'pad')
df["writer"] = df["writer"].fillna(method = 'pad')
df["star"] = df["star"].fillna(method = 'pad')
df["country"] = df["country"].fillna(method = 'pad')
df["company"] = df["company"].fillna(method = 'pad')
df["runtime"] = df["runtime"].fillna(method = 'pad')
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
df.head()
```

```python
# Converting the column "runtime" to interger

df["runtime"] = df["runtime"].astype("int64")
df["score"] = df["score"].astype("int64")
```

```python
df.head()
```

```python
# Create correct column 

df['released'] = df['released'].astype(str).str[:4]
```
